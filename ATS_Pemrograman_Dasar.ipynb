```python
try:
    kapasitas_maksimum = float(input("Masukkan kapasitas maksimum (kW/kVA): "))
    beban_terpakai = float(input("Masukkan beban terpakai (kW/kVA): "))

    if kapasitas_maksimum <= 0:
        print("Kapasitas maksimum harus lebih besar dari 0.")
    elif beban_terpakai < 0:
        print("Beban terpakai tidak boleh negatif.")
    else:
        persentase = (beban_terpakai / kapasitas_maksimum) * 100

        status_beban = ""
        if persentase <= 60:
            status_beban = "Beban Ringan"
        elif persentase <= 85:
            status_beban = "Beban Sedang"
        else:
            status_beban = "Beban Berat"

        print("\n--- HASIL KLASIFIKASI ---")
        print(f"Persentase Pemakaian: {persentase:.2f}%")
        print(f"Status Beban: **{status_beban}**")

except ValueError:
    print("Input tidak valid. Harap masukkan nilai numerik yang benar.")
```

```python
import random

# --- FUNGSI UTAMA ---
def hitung_efisiensi(daya_output, daya_input_matahari):
    """
    Menghitung efisiensi panel surya dalam persentase.
    Memastikan tidak ada pembagian dengan nol.
    """
    if daya_input_matahari > 0:
        efisiensi = (daya_output / daya_input_matahari) * 100
        return efisiensi
    else:
        return 0.0

def klasifikasi_efisiensi(efisiensi):
    """
    Menggunakan percabangan if/elif/else untuk mengklasifikasikan status efisiensi.
    """
    if efisiensi >= 18.0:
        return "**Optimal** (Efisiensi Tinggi)"
    elif 12.0 < efisiensi < 18.0:
        return "Normal (Efisiensi Sedang)"
    else:
        return "Perlu Perhatian (Efisiensi Rendah)"

def simulasi_harian(jam_operasional_max=8):
    """
    Simulasi pembacaan data panel surya selama jam-jam operasional utama.
    Menggunakan loop while untuk iterasi waktu.
    """
    data_efisiensi = []

    # Inisialisasi variabel untuk loop while
    jam_saat_ini = 8 # Mulai dari jam 8 pagi

    print("\n--- SIMULASI OPERASIONAL PANEL SURYA DIMULAI ---")

    # 1. PERULANGAN WHILE: Loop berjalan selama jam operasional belum melebihi batas (misalnya, jam 4 sore)
    while jam_saat_ini < (8 + jam_operasional_max):
        # Data simulasi: Daya yang diterima (input) dan daya yang dihasilkan (output)
        # Efisiensi di sini dipengaruhi oleh 'cuaca' (faktor acak)
        faktor_cuaca = random.uniform(0.8, 1.2) # Fluktuasi 80% - 120%
        daya_input = 800 * faktor_cuaca # Misal daya matahari 800 Watt/m²
        daya_output = daya_input * random.uniform(0.1, 0.22) # Output ~10% sampai 22% dari input

        # Hitung efisiensi menggunakan FUNGSI
        eff = hitung_efisiensi(daya_output, daya_input)

        data_efisiensi.append((jam_saat_ini, eff))

        print(f"[{jam_saat_ini}:00] Daya Input: {daya_input:.2f} W/m², Efisiensi: {eff:.2f}%")

        # Increment jam untuk iterasi berikutnya
        jam_saat_ini += 1

    print("--- SIMULASI SELESAI ---\n")
    return data_efisiensi

# --- PROGRAM UTAMA ---

# Jalankan simulasi harian untuk mendapatkan data
hasil_simulasi = simulasi_harian(jam_operasional_max=8)

## 2. Pengecekan dan Analisis Data Harian (PERULANGAN FOR dan PERCABANGAN IF/ELIF/ELSE)
print("--- ANALISIS DATA HARIAN ---")

total_efisiensi = 0
titik_warning = 0

# 2. PERULANGAN FOR: Iterasi melalui setiap data hasil simulasi
for jam, efisiensi in hasil_simulasi:

    # Klasifikasi status menggunakan FUNGSI
    status = klasifikasi_efisiensi(efisiensi)

    # Cek untuk Warning (PERCABANGAN IF/ELIF/ELSE diterapkan di dalam fungsi klasifikasi)
    if "Perlu Perhatian" in status:
        titik_warning += 1
        print(f"Peringatan di [{jam}:00]: Efisiensi hanya {efisiensi:.2f}% ({status}).")

    total_efisiensi += efisiensi

# Hitung rata-rata efisiensi harian
if hasil_simulasi:
    rata_rata_efisiensi = total_efisiensi / len(hasil_simulasi)
    status_rata_rata = klasifikasi_efisiensi(rata_rata_efisiensi)
else:
    rata_rata_efisiensi = 0
    status_rata_rata = "Data tidak ada"

print("\n--- RINGKASAN HARIAN ---")
print(f"Total {len(hasil_simulasi)} pembacaan data dilakukan.")
print(f"Rata-rata Efisiensi Harian: **{rata_rata_efisiensi:.2f}%**")
print(f"Status Umum Harian: {status_rata_rata}")
print(f"Jumlah Titik dengan Efisiensi Rendah: {titik_warning} kali")

# Final decision menggunakan IF/ELSE sederhana
if titik_warning > 2:
    print("\nKeputusan: Lakukan Pemeliharaan dan Pembersihan Panel Besok.")
else:
    print("\nKeputusan: Operasi Harian Normal, Lanjutkan Monitoring.")
```
